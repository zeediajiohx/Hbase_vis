```python
import sqlite3
import pandas as pd
```

```python
def create_table():
    conn = sqlite3.connect('population.db')
    print("Opened database successfully")
    c = conn.cursor()
    #世界疫情数据
    c.execute('''create table people
    (
        id int auto_increment
            primary key,
        year int null,
        country varchar(50) null ,
        continent varchar(50) null ,
        rank int null ,
        population int null 
    )''')
    print("Table created successfully")
    conn.close()
```

```python
create_table()
```

```python
def get_conn():
    #连接数据库，如果不存在则会在当前目录创建
    conn = sqlite3.connect("population.db")
    # c创建游标
    cursor = conn.cursor()
    return conn, cursor

#关闭连接
def close_conn(conn, cursor):
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```
